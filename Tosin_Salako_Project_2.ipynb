```python
#Renaming variable, Total Revenue
Labels = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
       'Order Date', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
       'Unit Cost', 'Total_Revenue', 'Total Cost', 'Total Profit','Date time']
```

```python
df.columns
```

```python
#To predict Total Revenue y = df['Total Revenue'] 
y = df.Total_Revenue
y
```

```python
#RFE Recursive Feature Elimination
#Predictive Power Score
#Correlation Matrix
#Univariate, bivariate and multivariate data analysis
#keyword argument and positional argument
```

what feature do I want to predict using deciion tree model?
which variables will affect it?

```python
Revenue_features = ['Units Sold', 'Unit Price']
```

```python
X = df[Revenue_features]
X
```

```python
from sklearn.tree import DecisionTreeRegressor

revenue_model = DecisionTreeRegressor(random_state=1)

revenue_model.fit(X,y)
```

```python
revenue_model.predict(X.head())
```

```python
#MY FIRST MACHINE MODEL
y
```

```python
print("Making predictions for Total_Revenue:")
print(X.head())
print("The predictions are")
print(revenue_model.predict(X.head()))
print('My ML Model:)')
```

#Model Evaluation _MAE

```python
from sklearn.metrics import mean_absolute_error

predicted_sales = revenue_model.predict(X)
mean_absolute_error(y, predicted_sales)
```
